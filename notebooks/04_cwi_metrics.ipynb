```python
from pathlib import Path
import pandas as pd

BASE = Path('..')
if not (BASE / 'data').exists():
    BASE = Path('.')

events = pd.read_csv(BASE / 'data/event_logs/synthetic_integrated_event_log.csv')
events['timestamp'] = pd.to_datetime(events['timestamp'])
events['exception_type'] = events['exception_type'].fillna('')
events.head()
```

```python
activity_counts = events['activity'].value_counts().rename_axis('activity').reset_index(name='count')
exception_counts = events[events['exception_type']!='']['exception_type'].value_counts().rename_axis('exception_type').reset_index(name='count')
activity_counts, exception_counts
```

```python
cycle = events.groupby('case_id')['timestamp'].agg(case_start='min', case_end='max').reset_index()
cycle['cycle_time_minutes'] = ((cycle['case_end'] - cycle['case_start']).dt.total_seconds()/60).round(1)
cycle
```
