```python
from pathlib import Path
import pandas as pd

BASE = Path('..')
if not (BASE / 'data').exists():
    BASE = Path('.')

df = pd.read_csv(BASE / 'data/event_logs/synthetic_integrated_event_log.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'])
variants = (
    df.sort_values(['case_id','timestamp'])
      .groupby('case_id')['activity']
      .apply(lambda s: ' > '.join(s))
      .reset_index(name='variant')
)
variants
```

```python
variant_counts = variants['variant'].value_counts().rename_axis('variant').reset_index(name='case_count')
variant_counts
```
